Write a helper function that finds the index in our embed dict that contains the given vector. Otherwise returns -1

```python
def get_id(embed, vec):
    if np.all(vec == 0):
        return -1
    assert vec.shape[-1] == embed.shape[-1], (vec.shape, embed.shape)
    element_comp = np.equal(embed, vec)
    row_comp = np.all(element_comp, axis=1)
    indices = np.argwhere(row_comp)
    assert indices.size == 1, f'onehot {onehot}\nindices {indices}\n'
    return indices.item()
```

### Sequence Generation

Now to write a dataloader that generates sequences from our new embedding dataset, which can be used to learn graph embeddings.

```python
import h5py
import numpy as np
from numpy.random import default_rng
```

```python
graphs = h5py.File('postera_protease2_id_graph.hdf5','r')
atom_seq_dataset = open('atom_sequences.csv', 'w')
descriptor_seq_dataset = open('descriptor_sequences.csv', 'w')
```

```python
def log_softmax(x):
    c = x.max()
    logsumexp = np.log(np.exp(x - c).sum())
    return x - c - logsumexp

# gets the neighbors of node within a given radius
# distance matrix should be a square matrix
# (optional) list of ban nodes which should not be considered
# returns the node IDs of the neighbors
# also returns the log probability vector based on their distance
def get_neighbors(node, distance_mat, radius, ban=[]):
    neighbors = np.nonzero(distance_mat[node, :] <= radius)[0]
    
    if neighbors.size and ban:
        conditions = sum([neighbors == x for x in ban])
        mask = np.nonzero(conditions == 0)
        neighbors = neighbors[mask]
        
    if neighbors.size == 0:
        return [], []
    
    distances = distance_mat[node, neighbors]
    log_probs = log_softmax(radius - distances)
    # increment neighbor ids because distance matrix excludes diagonal
    return neighbors, log_probs
```

```python
rng = default_rng()
k = 100

num_graphs = len(graphs.keys())
# generate k sequences per graph
for i, name in enumerate(graphs.keys()):
    coords = graphs[name]['coordinates'][()]
    atoms = graphs[name]['atoms'][()]
    descriptors = graphs[name]['descriptors'][()]
    
    # find and extract non-padding rows
    assert coords.ndim == 2, coords.shape
    rows = np.all(coords != 0, axis=1)
    indices = np.nonzero(rows)
    coords = coords[indices]
    atoms = atoms[indices]
    descriptors = descriptors[indices]
    
    # find good neighborhood radius (2 times the mean of shortest distances between atoms)
    # for distance_matrix calculation see pytorch documents for pdist
    # https://pytorch.org/docs/stable/generated/torch.nn.functional.pdist.html
    # diagonal is set to infinity 
    distance_matrix = np.linalg.norm((coords[:, None] - coords), axis=-1)
    distance_matrix[distance_matrix == 0] = np.inf
    radius = distance_matrix.min(axis=1).mean() * 2
    
    # generate random sequences
    for _ in range(k):
        # initial node to begin sequence
        node = rng.choice(indices[0])

        # initialize sequences to be written
        atom_seq = [atoms[node].item()]
        descriptor_seq = [descriptors[node].item()]

        # add node to ban list
        ban = [node]

        # get neighborhood and hop to a random neighbor
        neighbors, log_probs = get_neighbors(node, distance_matrix, radius, ban)
        while len(neighbors):
            node = rng.choice(neighbors, p=np.exp(log_probs))
            atom_seq.append(atoms[node].item())
            descriptor_seq.append(descriptors[node].item())
            ban.append(node)
            neighbors, log_probs = get_neighbors(node, distance_matrix, radius, ban)
        
        # save sequences to csv dataset
        # each line is a sequence
        atom_seq_dataset.write(','.join(map(str, atom_seq)) + '\n')
        descriptor_seq_dataset.write(','.join(map(str, descriptor_seq)) + '\n')
    
    print(f'{i}/{num_graphs}')
```

```python
graphs.close()
atom_seq_dataset.close()
descriptor_seq_dataset.close()
```
